## 2 .Si chiede di trovare, per ogni continente:
## a. il numero di casi fin dall'inizio della pandemia 
## b. la percentuale rispetto al totale mondiale del numero di casi

a. il numero di casi fin dall'inizio della pandemia 

```python
dati_nulli = covid[covid['continent'].notna()]
dati_continente = dati_nulli.groupby('continent')['new_cases'].sum()
print(dati_continente)
```

La riga sottostante dimostra che i dati presenti nella colonna 'location' corrispondono ai risultati della colonna 'continent'. In questo caso, abbiamo utilizzato 'Europe' come esempio, ma sostituendo con gli altri nomi presenti nella colonna 'continent', otteniamo risultati analoghi. Questo evidenzia una coerenza tra le due colonne, confermando che i valori della colonna 'location' riflettono correttamente i dati della colonna 'continent'

```python
location = covid['location']== 'Europe'
dati_location = covid[location]
totale_location = dati_location['new_cases'].sum()
print(f'il totale dei casi in Europa dall inizio della pandemia è : {totale_location}')
```

b. la percentuale rispetto al totale mondiale del numero di casi

```python
mondo = covid['location'] == 'World'
dati_mondiali = covid[mondo]
totale_mondiale = dati_mondiali['new_cases'].sum()
print(f'il totale mondiale dall inizio della pandemia è : {totale_mondiale}')
```

```python
percentuale_mondiale = (dati_continente/totale_mondiale)*100
percentuale_mondiale
```

## 3. Selezionare i dati relativi all'Italia nel 2022 e, poiché i nuovi casi vengono registrati settimanalmente, filtrare via i giorni che non hanno misurazioni; quindi mostrare con dei grafici adeguati: 
## a.l'evoluzione del casi totali dall'inizio alla fine dell'anno 
## b. il numero di nuovi casi rispetto alla data

Prima di iniziare a svolgere gli esercizi riportati precedentemente, ho modificato il tipo di dato (dtype) da 'object' a 'datetime'. Questa conversione consente di gestire i dati in modo più efficace, facilitando operazioni e analisi temporali sui dati. Utilizzando il tipo di dato 'datetime', possiamo eseguire operazioni come il filtraggio, il raggruppamento e la manipolazione delle date in modo più semplice e intuitivo

```python
covid['date']= pd.to_datetime(covid['date'], dayfirst = False)
covid.info()
```

```python
italia = covid['location']== 'Italy'
duemilaventidue = covid['date'].dt.year == 2022
italia_2022 = covid[italia & duemilaventidue]
print(italia_2022)
```

```python
italia_2022['total_cases'].isna().sum()
```

```python
italia_2022 = italia_2022.dropna(subset='total_cases')
italia_2022
```

```python
italia_2022.plot(figsize= (12,6),  x = 'date', y= 'total_cases', kind= 'line', title = 'andamento casi totali in Italia nel 2022', grid= True);
```

```python
italia_2022.plot(figsize = (12,6), x = 'date', y= 'new_cases', kind= 'line', title = 'andamento nuovi casi in Italia nel 2022', grid= True);
```

Per ottenere un grafico più lineare, ho applicato un filtro che include nella colonna 'new_cases' e nella colonna 'total_cases' solo i giorni in cui sono stati registrati nuovi casi, escludendo quindi quelli con valori pari a zero. Questo approccio permette di visualizzare in modo più chiaro le variazioni significative dei nuovi casi nel tempo.

```python
italia_2022_0 = italia_2022[italia_2022['new_cases'] != 0]
print(italia_2022_0)
```

```python
italia_2022_0.plot(figsize = (12,6), x = 'date', y= 'total_cases', kind= 'line', title = 'andamento casi totali in Italia nel 2022',grid= True);
```

```python
italia_2022_0.plot(figsize = (12,6),x = 'date', y= 'new_cases', kind= 'line',grid= True )
```

## 4 Riguardo le nazioni di Italia, Germania e Francia: 
## a.mostrare in un boxplot la differenza tra queste nazioni riguardo il numero di pazienti in terapia intensiva (Intensive Care Unit, ICU, considerare quindi la colonna icu_patients) da maggio 2022 (incluso) ad aprile 2023 (incluso)
## b.scrivere un breve commento (una o due righe) riguardo che conclusioni possiamo trarre osservando il grafico risultante

```python
nazioni_date = (covid['location'].isin(['Italy', 'Germany', 'France'])) & covid['date'].between('2022-05-01', '2023-04-30')
nazioni_date = covid.loc[nazioni_date, ['icu_patients', 'location', 'date']]
nazioni_date
```

```python
sns.boxplot(data = nazioni_date, x = 'location', y= 'icu_patients')
```

Il grafico evidenzia una chiara differenza nel numero di pazienti in terapia intensiva tra le tre nazioni, con Francia e Germania che mostrano valori più alti rispetto all'Italia. Nello specifico la Germania ha una mediana leggermente sopra i 1000 pazienti, indicando che ci sono molti pazienti in terapia intensiva, la Francia ha una mediana intorno a 1000 pazienti, suggerendo un numero relativamente elevato di pazienti in terapia intensiva e infine l'Italia è significativamente più basso, con una mediana che si avvicina a 250 pazienti e questo suggerisce una minore incidenza di pazienti in terapia intensiva rispetto alle altre due nazioni.

## 5. Riguardo le nazioni di Italia, Germania, Francia e Spagna in tutto il 2021: 
## a. mostrare, in maniera grafica oppure numerica, la somma dei pazienti ospitalizzati per ognuna (colonna hosp_patients)
## b. se ci sono dati nulli, con un breve commento scrivere se può essere possibile gestirli tramite sostituzione o meno

```python
nazioni_2021= (covid['location'].isin(['Italy', 'Germany', 'France', 'Spain'])) & (covid['date'].dt.year == 2021)
nazioni_2021 = covid.loc[nazioni_2021, ['location', 'hosp_patients']]
ragg_nazioni2021 = nazioni_2021.groupby('location')['hosp_patients'].sum()
ragg_nazioni2021
```

```python
ragg_nazioni2021.plot(kind='bar')
```

Analizzando il grafico, si può notare subito che in Germania non sono stati registrati dati ospedalieri. Pertanto, riempire la tabella con informazioni per questo paese non avrebbe senso, poiché inseriremmo dati privi di significato.Per quanto riguarda le altre nazioni, emerge chiaramente che la Francia ha registrato un numero elevato di casi ospedalieri rispetto agli altri paesi considerati. Al contrario, la Spagna ha riportato il valore più basso. L'Italia, invece, si colloca in una posizione intermedia tra la Francia e la Spagna.
